```python
import recordlinkage
import pandas as pd
import os
```

```python
#path = os.path.abspath(r"D:\uni\magister\Semantic Web\project\code")
df1 = pd.read_csv("query_result_municipal.csv")
df2 = pd.read_csv("query_result_rest.csv")

df = pd.concat([df1, df2])
len(df)
```

```python
df.insert(0, 'id', range(0, 0 + len(df)))
df.set_index('id', inplace = True)

df = df[["s", "election_type", "name", "local_party_label", "main_party_label", "mun_label", "mir_label", "main_el_label", "elLabel"]]
```

```python
preprocessed_df = df.copy()

string_columns = ["name", "local_party_label", "main_party_label", "mun_label", "mir_label", "main_el_label", "elLabel"]
for column_label in string_columns:
    preprocessed_df[column_label] = preprocessed_df[column_label].str.lower()

preprocessed_df.describe()
```

```python
preprocessed_df["name"].value_counts()
```

```python
indexer = recordlinkage.Index()
indexer.block("name")

candidate_links = indexer.index(preprocessed_df)
len(candidate_links)
```

```python
comparer = recordlinkage.Compare()
#comparer.exact("s", "s", label="s")
comparer.exact("local_party_label", "local_party_label", label="local_party_label")
comparer.exact("main_party_label", "main_party_label", label="main_party_label")
comparer.exact("mun_label", "mun_label", label="mun_label")
comparer.exact("mir_label", "mir_label", label="mir_label")

features = comparer.compute(candidate_links, preprocessed_df)
features.sum(axis=1).value_counts().sort_index(ascending=False)
```

```python
#features["s"] = features["s"] * -4
features["local_party_label"] = features["local_party_label"] * 0.75
features["main_party_label"] = features["main_party_label"] * 0.5
features["mun_label"] = features["mun_label"] * 1.5
features["mir_label"] = features["mir_label"] * 1.25

matches = features[features.sum(axis=1) >= 1.0]
matches
```

```python
grouped = matches.reset_index().groupby("id_1")
len(grouped)
```

```python
grouped.size().value_counts()
```

```python
def get_formatted_record(record):
    return (
        f'{record["s"]}, '
        f'{record["election_type"]}, '
        f'{record["local_party_label"]}, '
        f'{record["main_party_label"]}, '
        f'{record["mun_label"]}, '
        f'{record["mir_label"]}, '
        f'{record["main_el_label"]}, '
        f'{record["elLabel"]}\n'
    )

def get_matches(df, index, matched_with_indices):
    item = df.loc[index]
    matched_items = [df.loc[i] for i in matched_with_indices]
    
    return (item, matched_items)


with open("output.txt", "w", encoding="utf-8") as output:
    output.write("s, election_type, local_party_label, main_party_label, mun_label, mir_label, main_el_label, elLabel\n\n")
    
    i = len(grouped)
    for name, group in grouped:
        record, matched_with = get_matches(df, name, group["id_2"].values)
            
        output.write(f'"{record["name"]}" matched with {len(matched_with) + 1} records:\n')
        output.write(get_formatted_record(record))
        for match in matched_with:
            output.write(get_formatted_record(match))
        output.write("------------------------------------------------------------\n")
        
        i = i - 1
        print(f"{i} left", end='\r')
        
```
